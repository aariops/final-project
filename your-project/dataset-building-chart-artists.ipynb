```python
def get_artist_cred(x):
    if type(x) == dict:
        return x.get("artist-credit")
    
def get_artist(x):
    if type(x) == dict:
        return x.get("name")

def get_disc_id(x):
    if type(x) == dict:
        return x.get("id")
    
def get_title(x):
    if type(x) == dict:
        return x.get("title")

def get_date(x):
    if type(x) == dict:
        return x.get("date")
```

```python
new_df["artist_cred"] = new_df["release-list"].apply(get_artist_cred)
new_df["release_id"] = new_df["release-list"].apply(get_disc_id)
new_df["album"] = new_df["release-list"].apply(get_title)
new_df["year"] = new_df["release-list"].apply(get_date)
```
